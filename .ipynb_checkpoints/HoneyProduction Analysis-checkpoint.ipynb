## graphical representation of relation of year and totalprod

```python
sns.relplot(x='year',y='totalprod',data = df, kind = 'line')
```

```python
ptfig = plt.figure(figsize=(15,7))
sns.pointplot(x='year',y='totalprod',data = df, kind = 'line')
ptfig.savefig('totalprodvsYear.png')
```

```python
nTfig = plt.figure(figsize=(25,7))
sns.relplot(x='numcol',y='totalprod',data = df)
```

```python
df[['numcol','totalprod']].corr()   #corr of 0.95 +ve indicates almost linear increase of totalprod with increase in numcol
```

## what is the correlation between all the numerical values

```python
sns.pairplot(data= df[['numcol','yieldpercol','totalprod','stocks',
                      'priceperlb','prodvalue','year']] )
```

```python
cor = df[['numcol','yieldpercol','totalprod','stocks',
                      'priceperlb','prodvalue','year']].corr()
cor
```

```python
sns.heatmap(cor,annot=True, cmap='plasma',vmin = -1,  vmax= 1)
```

## production trend from 1998 to 2012?

```python
sns.pointplot(x= 'year',y= 'totalprod',data= df)  #gives better view than line kind as 
#for example, in 2010 the total prod has jumped to high value
```

## production trend for each state?

```python
sns.relplot(x= 'year',y = 'totalprod',data = df,kind= 'line',hue='state')
```

```python
#instead of above graph we could use FaceGrid and column state to determine the relation
```

```python
eachstateProdTrend = plt.figure(figsize=(10,5))
g =  sns.FacetGrid(df, col= 'state', col_wrap=5 , height = 10)
g = g.map(plt.plot, 'year','totalprod',marker = ".")
eachstateProdTrend.savefig('stateProdtrend.png')
```

```python
#from above we could see SD state production overall increased
```

## linear regression - finding linear relationship

```python
sns.lmplot(x='stocks',y='prodvalue',data = df)
```
