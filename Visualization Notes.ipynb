```python
import seaborn as sns
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x = np.arange(100)
y = x**2
```

```python
quadfig = plt.figure(figsize=(12,5))
plt.plot(x,y, 'r')
plt.title('Quad Function')
plt.xlabel('x')
plt.ylabel('y= x^2')
quadfig.savefig('Quad.png')
#plt.close()
```

```python
df = sns.load_dataset('flights')
df.head()
```

```python
sns.relplot(x='passengers',y='month', data = df)  #default is scatter plot
```

```python
sns.relplot(x='passengers',y='month', hue='year', data = df)
```
